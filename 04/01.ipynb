```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

while True:
    command = input('Введите команду: ')

    if command == 'p':
        doc_number = input('Введите номер документа: ')
        name = None

        for i in documents:
            if doc_number == i['number']:
                name = i['name']

        if name is not None:
            print(f'Владелец документа: {name}')
        else:
            print('Результат:\nДокумент не найден в базе')

    if command == 's':
        doc_number = input('Введите номер документа: ')
        result = None

        for key, value in zip(directories.keys(), directories.values()):
            if doc_number in value:
                print(f'Документ хранится на полке: {key}')

    if command == 'l':

        for i in range(len(documents)):
            print(f"№: {documents[i]['number']}, тип: {documents[i]['type']}, владелец: {documents[i]['name']}, полка "
                  f"хранения: {i}")

    if command == 'ads':
        number = input('Введите номер полки: ')

        if number in directories.keys():
            print('Такая полка уже существует. Текущий перечень полок: ' + ', '.join(str(i) for i in directories.keys()))
        else:
            directories.update({number: []})
            print('Полка добавлена. Текущий перечень полок: ' + ', '.join(str(i) for i in directories.keys()))

    if command == 'ds':
        number = input('Введите номер полки: ')

        if number in directories.keys() and len(directories.get(number)) > 0:
            print(
                'На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: ' + ', '.join(str(i) for i in directories.keys()))
        else:
            if len(directories.get(number)) == 0:
                directories.pop(number)
                print('Полка удалена. Текущий перечень полок: ' + ', '.join(str(i) for i in directories.keys()))

    if command == 'q':
        break
```
